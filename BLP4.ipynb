```python
import os
from openai import OpenAI

from google.colab import userdata

api_key=userdata.get('OPENAI_API_KEY')

client = OpenAI(api_key=api_key)
```

```python
prompt_system = """ You are a helpful assistant whose goal is to write short poems."""

prompt = """Write a short poem about summer."""
response = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {"role": "system", "content": prompt_system},
    {"role": "user", "content": prompt}
  ],
)

print(response.choices[0].message.content)
```

```python
prompt1 = "What is the name of the famous scientist who developed the theory of general relativity?"

prompt2 = "Provide a brief description of that scientist's theory of general relativity."

response1 = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {"role": "user", "content": prompt1}
  ]
)

scientist_name= response1.choices[0].message.content.strip()

prompt2 = f"""Provide a brief description of {scientist_name}'s theory of general relativity."""

response2 = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {"role": "user", "content": prompt2}
  ]
)

theory_desc= response2.choices[0].message.content.strip()
print(f"Scientist: {scientist_name}")
print(f"Theory of General Relativity: {theory_desc}")
```

```python
#Example 2:
from transformers import pipeline

def summarize_reviews(reviews):
    summarizer = pipeline("summarization")
    summary = summarizer(reviews, max_length=50, min_length=20, do_sample=False)
    return summary[0]['summary_text']


reviews = """This product is amazing! I love the build quality and the battery life.
However, the software needs improvement. The UI is a bit clunky."""
print(summarize_reviews(reviews))
```

```python
#Example 1
import openai
import os

# Set up OpenAI API Key
OPENAI_API_KEY = "OPENAI_API_KEY"
client = openai.Client(api_key=OPENAI_API_KEY)

def generate_email(subject, recipient, context):
    prompt = f"""
    Write a professional email with the following details:
    Subject: {subject}
    Recipient: {recipient}
    Context: {context}
    """
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "You are a helpful assistant."},
                  {"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content


subject = "Project Update Request"
recipient = "John Doe"
context = "Requesting a status update on the ABC project."
print(generate_email(subject, recipient, context))
```

```python
# Example 3: Creative Ad Copy Generation
def generate_ad_copy(product_name, key_features):
    prompt = f"""
    Generate a catchy advertisement for {product_name} highlighting its features: {key_features}.
    Make it engaging and persuasive.
    """
    response = client.chat.completions.create(
        model="gpt-4",
        messages=[{"role": "system", "content": "You are a creative marketing expert."},
                  {"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content

# Example usage
product_name = "SmartWatch X"
key_features = "Heart Rate Monitoring, Long Battery Life, Stylish Design"
print(generate_ad_copy(product_name, key_features))
```
